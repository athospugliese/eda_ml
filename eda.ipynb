```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualização 1: Distribuição da idade dos indivíduos
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Distribuição da Idade')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
# Importando as bibliotecas necessárias
import matplotlib.pyplot as plt

# Plotando o gráfico de barras
df['employment_status'].value_counts().plot(kind='bar', figsize=(10,5))

# Adicionando título e rótulos aos eixos
plt.title("Número de indivíduos por status de emprego")
plt.ylabel('Número de indivíduos')
plt.xlabel('Status de Emprego')

# Exibindo o gráfico
plt.show()
```

```python
df['employment_status'].value_counts().nlargest(10).plot(kind='bar', figsize=(10,5))

plt.title("Os 10 empregos mais comuns")
plt.ylabel('Número de indivíduos')
plt.xlabel('Status de Emprego')

plt.show()
```
